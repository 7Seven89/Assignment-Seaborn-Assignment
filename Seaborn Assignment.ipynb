```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

**Q1. Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.**

A1.

**Scatter Plot (sns.scatterplot):**
- Use: Visualizes the relationship between two continuous variables. It is helpful to observe correlations, trends, or outliers in the data.
- Example: Comparing height vs. weight of individuals.

**Line Plot (sns.lineplot):**
- Use: Displays trends or changes over time. This is useful for time-series data or continuous data trends.
- Example: Plotting sales data over months to observe seasonality or growth.

**Bar Plot (sns.barplot):**
- Use: Shows the comparison of categorical data with their corresponding values. The height of the bars represents the value.
- Example: Comparing sales figures across different regions or products.

**Box Plot (sns.boxplot):**
- Use: Visualizes the distribution of a dataset, highlighting the median, quartiles, and potential outliers. It is useful for understanding the spread and skewness of data.
- Example: Showing the distribution of exam scores across different classes.

**Heatmap (sns.heatmap):**
- Use: Provides a graphical representation of data where values are represented by color intensity. Often used for correlation matrices or displaying two-dimensional data.
- Example: Visualizing the correlation between different variables in a dataset.

**Q2. Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.**

```python
df = sns.load_dataset('fmri')
```

```python
df
```

```python
plt.figure(figsize= (10, 6))
sns.lineplot(data = df, x= 'timepoint', y= 'signal', hue= 'event', style= 'region')
plt.show()
```

**Q3. Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.**

```python
df1 = sns.load_dataset('titanic')
```

```python
df1
```

```python
plt.figure(figsize= (14, 6))

for i, col in enumerate(['age', 'fare']):
    plt.subplot(1, 2, i+1)
    sns.boxplot(data= df1, x= 'pclass', y= col)

plt.tight_layout()
plt.show()
```

**Q6. Use the "flights" dataset from seaborn to plot a heatmap.**

```python
df5 = sns.load_dataset('flights')
```

```python
df5
```

```python
df5['month'] = pd.to_datetime(df5['month'], format= '%b').dt.month
```

```python
corr_mat = df5.corr()
```

```python
plt.figure(figsize= (10,6))
sns.heatmap(data= corr_mat, annot= True, cmap= 'coolwarm')
plt.show()
```
